```python
import cv2
 
from multiprocessing import pool
 
from multiprocessing.dummy import Pool as ThreadPool
import os 



# This function loads a file, resize it and write in the output folder
 
def create_icon( inputFileName, resize_shape = (224,224) ):
    '''resizes images inplace'''
    
    im = cv2.imread( inputFileName )

    small_im = cv2.resize( im, resize_shape )

    cv2.imwrite( inputFileName, small_im )
 


 
# Load the list of images in the array lines
```

```python
from pathlib2 import Path
```

```python
path = Path('data/jpeg/train-224/')
```

```python
imagesList = list(path.glob('*.jpg'))
```

```python
imagesList =  [elem.as_posix() for elem in imagesList]
```

```python
pool = ThreadPool(10)
 
pool.map( create_icon, imagesList )
```
